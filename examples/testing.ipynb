```python
import tapsap
import numpy as np
import pandas as pd

data = pd.read_csv('~/Desktop/surface_data_ross/ross_surface_1/results_shifted_2_folder/flux_data/CH4.csv', header=None)
times = np.array(data[0])
inert = data[1]

data = pd.read_csv('~/Desktop/surface_data_ross/ross_surface_2/results_shifted_2_folder/flux_data/CH4.csv', header=None)
flux = data[1]

data = pd.read_csv('~/Desktop/surface_data_ross/ross_surface_2/results_shifted_2_folder/thin_data/CH4.csv', header=None)
concentration = data[0]

data = pd.read_csv('~/Desktop/surface_data_ross/ross_surface_2/results_shifted_2_folder/thin_data/r_CH4.csv', header=None)
rate = data[0]

zone_lengths = {'zone0':2, 'zone1' : 0.12, 'zone2':2}
```

```python
inert_max = np.max(inert)
df = pd.DataFrame()
df['inert'] = inert / np.max(inert)
df['flux'] = flux / np.max(inert)
df['pre_rate'] = df['inert'] - df['flux']
df['pre_rate'] /= np.max(df['pre_rate'])
df['concentration'] = concentration / np.max(concentration)
df['rate'] = rate / np.max(rate)
tapsap.plot_tap(times, df, legend=True)
```

```python
import math
def smooth_early(flux, times):
    peak_rt = np.argmax(flux)
    sub_times = times[1:(peak_rt + 1)]
    time_range = times[peak_rt]
    temp_mean = time_range / 2 
    temp_sd = temp_mean / 3
    temp_cdf = np.array([0.5 * (1 + math.erf((i - temp_mean) / (temp_sd * np.sqrt(2)))) for i in sub_times])
    flux[1:(peak_rt +1 )] = temp_cdf
    return flux
```
